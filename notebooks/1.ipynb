```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
agents = 20
tp_data = pd.read_csv(f'../tests/03/TP_a{agents}_t{agents}.csv', delimiter=';')
tp_data = list(tp_data['complete_time'])

tpts_data = pd.read_csv(f'../tests/03/TPTS_a{agents}_t{agents}.csv', delimiter=';')
tpts_data = list(tpts_data['complete_time'])

central_data = pd.read_csv(f'../tests/03/Central_a{agents}_t{agents}.csv', delimiter=';')
central_data = list(central_data['complete_time'])
```

```python
tp_out20 = [0]
for i in range(0, 600):
    curr = tp_out20[-1]
    while len(tp_data) > 0 and tp_data[0] == i:
        curr += 1
        tp_data.pop(0)
    tp_out20.append(curr)
tp_out20.pop(0)

tpts_out20 = [0]
for i in range(0, 600):
    curr = tpts_out20[-1]
    while len(tpts_data) > 0 and tpts_data[0] == i:
        curr += 1
        tpts_data.pop(0)
    tpts_out20.append(curr)
tpts_out20.pop(0)

central_out20 = [0]
for i in range(0, 600):
    curr = central_out20[-1]
    while len(central_data) > 0 and central_data[0] == i:
        curr += 1
        central_data.pop(0)
    central_out20.append(curr)
central_out20.pop(0)
```

```python
df = pd.DataFrame.from_dict({'TP-10':tp_out10, 'TPTS-10': tpts_out10, 'LCPD-10':central_out10,
                            'TP-20':tp_out20, 'TPTS-20': tpts_out20, 'LCPD-20':central_out20,
                            'TP-30':tp_out30, 'TPTS-30': tpts_out30, 'LCPD-30':central_out30})
```
